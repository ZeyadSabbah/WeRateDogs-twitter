# Visualization
## Rating Score Boxplot

It shows that most of the ratings lie between 1 and 1.2 with a median 1.1, and 1.5 is the maximum score.

## Day of Week Interactions
* Using line plot  

Retweet and Favourite counts are taken as an indication to users's interactions through the week. Line plot is the best plot showing the slope, it is easily noticed that both lines have the same graphs.  
Wednesday and Saturday are the peaks of interactions.  
The lines represents the mean of every day of the weak across the 1.7 years.  
* Using bar plot  

In order to compare between the Retweets and Favourites, bar plot does a better job here. It shows that users favourite tweets frequently rather than retweeting them.  

## Breeds Interactions
* Using horizontal bar plot
### Retweet Count

This shows the breeds which people interact with the most, Lakeland Terrier comes first in line with almost 14,500 retweets in average, and secondly comes Standard Poodle with around 12,000 retweets in average.  

### Favourite Count

This shows that Giant Lakeland Terrier still has the highest favourites with around 50,000 favourites in average, and secondly come French Bulldog and Standard Poodle with around 30,000 favourites in average.

## Tweets - Days

This gives an idea on when WeRateDogs tweets, there are not much of a difference between days, but it seems that it's decreasing through weekdays starting from Monday till it reaches the lowest value on Sunday.

## Scoring Rate Distribution

This graph shows the distribution, just like the boxplot, but in a simpler way showing the counts on the axis of each rating score to give sense of the dataset.
